## __1. Data Cleaning__

```python
df.isna().sum()
```

#### dapat di lihat terdapat beberapa kolom yang memiliki Nan Value Data

```python
df.drop(columns=['title_orig','merchant_name','merchant_info_subtitle','merchant_id',
                   'merchant_profile_picture','has_urgency_banner','urgency_text','product_url','product_picture','product_id','crawl_month','theme','currency_buyer'],inplace=True)
```

```python
df.info()
```

#### menghapus kolom yang tidak diperlukan untuk analisa lebih lanjut, seperti kolom has_urgency_banner, urgency_text dan merchant_profile_picture yang memiliki banyak missing Value.

```python
nan_replace={'has_urgency_banner':0,'urgency_text':'N/A','origin_country':'unknown','product_color':'unknown'}
df.fillna(nan_replace,inplace=True)
df_cln=df.dropna()
df_cln.info()
```
